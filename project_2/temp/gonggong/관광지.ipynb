```python
df_4=pd.read_csv('제주특별자치도_제주도장소(POI)데이터_20151231.csv',encoding='cp949')
df_4.head(2)
```

```python
df_4=df_4.drop([df_4.columns[0],df_4.columns[-1]],axis=1)
```

```python
df_4=df_4.rename(columns={df_4.columns[0]:'lng',df_4.columns[1]:'lat'})
```

```python
jj_reg=['한림읍','애월읍','구좌읍','조천읍','한경면','추자면','우도면']
seo_reg=['대정읍','남원읍','성산읍','안덕면','표선면']

reg_lst=[]

for addr in tqdm(df_4['소재지']):
    if addr.find('제주시')==-1:
        if addr.split(' ')[2] in seo_reg:
            reg_lst.append('제주특별자치도 서귀포시 '+addr.split(' ')[2])
        else:
            reg_lst.append('제주특별자치도 서귀포시')
    else:
        if addr.split(' ')[2] in jj_reg:
            reg_lst.append('제주특별자치도 제주시 '+addr.split(' ')[2])
        else:
            reg_lst.append('제주특별자치도 제주시')
```

```python
df_4['소재지'] = reg_lst
```

- 구두점 및 쓸데없는 문자 제거

```python
from nltk.corpus import stopwords
from string import punctuation

list_stopwords = set(stopwords.words('english') + list(punctuation))

df_1['관광지명']=df_1['관광지명'].apply(lambda x: [word for word in x if word not in list_stopwords])
df_1['관광지명']=df_1['관광지명'].apply(lambda x:('').join(x))
df_1['관광지명']=df_1['관광지명'].apply(lambda x:x.replace(' ',''))

df_2['관광지명']=df_2['관광지명'].apply(lambda x: [word for word in x if word not in list_stopwords])
df_2['관광지명']=df_2['관광지명'].apply(lambda x:('').join(x))
df_2['관광지명']=df_2['관광지명'].apply(lambda x:x.replace(' ',''))

df_35['관광지명']=df_35['관광지명'].apply(lambda x: [word for word in x if word not in list_stopwords])
df_35['관광지명']=df_35['관광지명'].apply(lambda x:('').join(x))
df_35['관광지명']=df_35['관광지명'].apply(lambda x:x.replace(' ',''))

df_4['장소명']=df_4['장소명'].apply(lambda x: [word for word in x if word not in list_stopwords])
df_4['장소명']=df_4['장소명'].apply(lambda x:('').join(x))
df_4['장소명']=df_4['장소명'].apply(lambda x:x.replace(' ',''))
```

```python
display(df_1.head(2),df_2.head(2),df_35.head(2),df_4.head(2))
print(df_1.shape,df_2.shape,df_35.shape,df_4.shape)
```
