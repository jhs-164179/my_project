## 제주시 도로 침수 우려지역

```python
import requests
from bs4 import BeautifulSoup

URL= 'https://bangjae.jeju.go.kr/ctrltower/state/flooding/road.htm'
res = requests.get(URL)
soup = BeautifulSoup(res.text, 'html.parser')
tags = soup.find_all('tr',attrs={'class':''})

for i, map in enumerate(tags):
    print('{1:}'.format(i, map.get_text()))
```
